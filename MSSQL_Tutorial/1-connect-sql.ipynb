```python
import pyodbc

DRIVER_NAME = 'ODBC Driver 17 for SQL Server'
SERVER_NAME = r'LAPTOP-D1GHCN4R\POWERBIREPORT'
DATABASE_NAME = 'AdventureWorksLT2014'
USERNAME = 'pbi'
PASSWORD = '123'

connection_string = f"""
DRIVER={{{DRIVER_NAME}}};
SERVER={SERVER_NAME};
DATABASE={DATABASE_NAME};
UID={USERNAME};
PWD={PASSWORD};
"""

conn = pyodbc.connect(connection_string)
print(conn)
```

```python
cursor = conn.cursor()
```

```python
# Read data
cursor.execute("""SELECT CustomerID
  FROM [AdventureWorksLT2014].[SalesLT].[Customer]""")
rows = cursor.fetchall()

# Print data
for row in rows:
    print(row)
```

## Close the connection

```python
conn.close()
```
